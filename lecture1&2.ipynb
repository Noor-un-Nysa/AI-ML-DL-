```python
#Operations and operands
#I've typecasted the input into integer as the input it was taking was in string.
var_num1=int(input("Enter your 1st number: "))
var_num2=int(input("Enter your 2nd number: "))
var_num3=0
#Lets just do basic D.M.A.S operations
var_num3=int(var_num1)/int(var_num2)
print(f"The division of {var_num1} and {var_num2} is {var_num3}")
var_num3=int(var_num1)*int(var_num2)
print(f"The multiplication of {var_num1} and {var_num2} is {var_num3}")
var_num3=int(var_num1)+int(var_num2)
print(f"The sum of {var_num1} and {var_num2} is {var_num3}")
var_num3=int(var_num1)-int(var_num2)
print(f"The subtraction of {var_num1} and {var_num2} is {var_num3}")

```

```python
#list using list()

# create list with the range function
list2=list(range(20))
print(list2)
#add an item
list2.append(21)
print(list2)
#removes an item but takes index by default last index
list2.pop()
print(list2)
#removes exact number/element
list2.remove(19)
print(list2)
# returns the specfic item by index
print(list2[4])
#replacing number
list2[2]=11
print(list2)
#insert func, it's 1st parameter is index and 2nd is value to be inserted
list2.insert(0,00)
print(list2)
```
