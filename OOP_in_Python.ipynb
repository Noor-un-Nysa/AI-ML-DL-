# Mini_Project: Bank Account System

---



Design a class BankAccount with these requirements:

Each account has account_holder and balance (instance variables).

The bank has a class variable bank_name (same for all accounts).

Create methods:

deposit(amount) → increases balance.

withdraw(amount) → decreases balance (but check for insufficient funds).

show_balance() → displays current balance.

Create a class method change_bank_name(new_name) to update the bank’s name.

Use a constructor to initialize account details, and a destructor to display a goodbye message when account is closed.

**Challenge:** Handle the condition where a withdrawal is greater than balance.


```python
class BankAccount:

  def __init__(self,acc_holder,bal,bnk):
    self.account=acc_holder
    self.balance=bal
    self.bank_name=bnk
    print(f"Account created under name of {self.account} with balance of {self.balance} in {self.bank_name}")
  def deposit(self,amount):
    self.balance=self.balance + amount
  def withdraw(self,amount):
    if self.balance>amount:
      self.balance=self.balance - amount
    else:
      print("Oppsie You got insuffient balance")

  def showbalance(self):
    print("Your current balance is: Rs ",self.balance,"/-")
  def changebankname(self, newname):
    self.bank_name=newname
    print("Your new bank name has changed to," , self.bank_name)
  def __del__(self):
    print("Thank you for banking with us")
#menu
print("Welcome to Account Creation")
print("------------------------------")
print("Press 1 to create account")
print("Press 2 change your Bank")
print("Press 3 to show balance")
print("Press 4 to deposit in account")
print("Press 5 to withdraw from account")
print("Press 6 to exit")
choice=int(input("Enter your choice: "))
if choice==1:
  username=input("Enter your Name: ")
  intial_balance=int(input("Enter your intial balance:Rs "))
  bank_name=input("Enter your bank name: ")
  c1=BankAccount(username,intial_balance,bank_name)
elif choice==2:
  bank_name=input("Enter your bank name: ")
  c1.changebankname(bank_name)
elif choice==3:
  c1.showbalance()
elif choice==4:
  amount=int(input("Enter the amount you want to deposit:Rs "))
  c1.deposit(amount)
  c1.showbalance()
elif choice==5:
  amount=int(input("Enter the amount you want to withdraw:Rs "))
  c1.withdraw(amount)
  c1.showbalance()
elif choice==6:
  del c1
else :
  print("Invalid choice")
```
