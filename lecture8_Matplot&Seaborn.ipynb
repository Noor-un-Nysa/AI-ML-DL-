```python
import matplotlib.pyplot as plt
import seaborn as sns
```

1. **Total Bill Distribution**:
   - **Question**: Create a histogram to visualize the distribution of `total_bill`.
   - **Task**: Customize the number of bins and add labels to the axes and a title. Discuss the shape of the distribution.

```python
sns.displot(data=df,x='total_bill',bins =20,kde=False)
plt.title('Total Bill Distribution')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
```


2. **Tip vs. Total Bill**:
   - **Question**: Create a scatter plot to visualize the relationship between `total_bill` and `tip`.
   - **Task**: Use different colors to represent the `sex` of the payer. Discuss any patterns observed.

```python
sns.scatterplot(data=df,x='total_bill',y='tip',hue='sex')
```


3. **Average Tip by Gender**:
   - **Question**: Create a bar plot to show the average tip amount given by `sex`.
   - **Task**: Annotate the bars with the exact average tip values.

```python
average_tip=df.groupby('sex')['tip'].mean()
sns.barplot(x=average_tip.index,y=average_tip.values)

# Add title and labels
plt.title("Average Tip Amount by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Tip (Rs)")
```


4. **Total Bill by Day**:
   - **Question**: Generate a box plot to visualize the distribution of `total_bill` for each day of the week.
   - **Task**: Discuss how the total bill varies by day and any potential outliers.

```python
sns.boxplot(data=df,x='total_bill',y='day')
#the bubbles are outliers
```


5. **Size of Party vs. Total Bill**:
   - **Question**: Create a scatter plot to examine the relationship between `size` (size of the party) and `total_bill`.
   - **Task**: Use different marker shapes or colors to indicate whether the party was a `smoker` or `non-smoker`.

```python
sns.scatterplot(data=df,x='size',y='total_bill',hue='smoker')
plt.xlabel('Size of Party')
plt.ylabel('Total Bill in Rupees')
```


6. **Price Per Person Analysis**:
   - **Question**: Create a violin plot to visualize the distribution of `price_per_person` across different `day`s.
   - **Task**: Discuss the differences in spending per person by day.

```python
sns.violinplot(data=df,x='day',y='price_per_person')
```


7. **Tips by Time of Day**:
   - **Question**: Create a box plot to compare the distribution of `tip` amounts for `Dinner` and `Lunch`.
   - **Task**: Discuss any differences in tipping behavior based on the time of day

```python
d_tip=df.groupby('time')['tip'].mean()

sns.barplot(x=d_tip.index,y=d_tip.values)
```


8. **Correlation Analysis**:
   - **Question**: Generate a heatmap to visualize the correlation between numerical variables in the dataset (e.g., `total_bill`, `tip`, `size`, `price_per_person`).
   - **Task**: Discuss any strong correlations found between the variables.

```python
numeric_cols = ['total_bill', 'tip', 'size', 'price_per_person']
sns.heatmap(df[numeric_cols].corr() )
```


9. **Smoker vs. Non-Smoker Spending**:
   - **Question**: Create a bar plot showing the average `total_bill` for smokers vs. non-smokers.
   - **Task**: Discuss how smoking status might influence spending.

```python
bill=df.groupby('smoker')['total_bill'].mean()
sns.barplot(x=bill.index,y=bill.values)
```


10. **Pie Chart of Payment Methods**:
