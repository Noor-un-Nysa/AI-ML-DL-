```python
print("Welcome to SIFT")

while True:   # keep program running
    # ---- Enter student data ----
    cnic_input = int(input("Enter CNIC: "))
    cnic_list = []
    cnic_list.append(cnic_input)

    student_marks = []
    english_marks = int(input("Enter Marks of English: "))
    student_marks.append(english_marks)
    maths_marks = int(input("Enter Marks of Maths: "))
    student_marks.append(maths_marks)
    urdu_marks = int(input("Enter Marks of Urdu: "))
    student_marks.append(urdu_marks)

    # ---- Menu ----
    print("\nWhat do you want to do?")
    print("Press 1 to Total")
    print("Press 2 to Average")
    print("Press 3 to Grade")
    print("Press 4 to Continue Entering Marks")
    print("Press 5 to Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        print("Total Marks =", sum(student_marks))

    elif choice == 2:
        avg = sum(student_marks) / len(student_marks)
        print("Average Marks =", avg)

    elif choice == 3:
        avg = sum(student_marks) / len(student_marks)
        if avg >= 90:
            print("Grade: A+")
        elif avg >= 80:
            print("Grade: A")
        elif avg >= 70:
            print("Grade: B")
        elif avg >= 60:
            print("Grade: C")
        else:
            print("Grade: Fail")

    elif choice == 4:
        continue   # goes back to top, new student entry

    elif choice == 5:
        print("Exiting program. Goodbye!")
        break

    else:
        print("Invalid choice, try again!")
```
