Task 2: Writing User Input to a File

Objective: Create a Python program that takes user input and writes it to a new file.

Instructions:

* Ask the user to enter a few lines of text.
* Open a file named user_input.txt in write mode.
* Write the user's input to the file.
* Inform the user that the input has been saved.

Sample Output:
```
Please enter your text (type 'STOP' to end):
Hello
This is a test.
STOP
Your input has been saved to user_input.txt
```

````python
fname="user.txt"
with open(fname,'w+') as mfile:
  while True:
    content=input("Enter your text")
    if content=="STOP":
      break
    else:
      mfile.write(content)
      mfile.seek(0)
      print(mfile.read())
print("Your input has been saved to user_input.txt")

````

Objective: Write a Python program that deletes a specific line from a file.

Instructions:

* Open delete_line.txt in read mode.
* Ask the user to enter the line number to delete.
* Create a new file updated_file.txt with the line removed.
* Replace the original file with the updated one.

````python
with open("userx.txt",'r') as x:
  inputx=int(input("Enter the line number to delete"))
  content=list(x.readlines())
  x.seek(0)
  with open("user.txt",'w+') as y:
    size= len(content)
    for i in range(size):
      if(i!=inputx):
        y.write(content[i])
    print(y.read())
````
