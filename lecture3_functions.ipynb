**MAP AND FILTER:**


```python

#numbers = [1, 2, 3, 4, 5]#using list
numbers = (1, 2, 3, 4, 5)#using tuples
#numbers = {1, 2, 3, 4, 5}#using sets

# using map with a function
def square(x):
    return x ** 2
squares = tuple(map(square, numbers))
print(squares)   # [1, 4, 9, 16, 25]
```

```python
#Example 2: Using lambda
numbers = [1, 2, 3, 4, 5]
squares = list(map(lambda x: x**2, numbers))
print(squares)   # [1, 4, 9, 16, 25]
```

```python
# Example 1: Keep even numbers
numbers = [1, 2, 3, 4, 5, 6]

def is_even(x):
    return x % 2 == 0

evens = list(filter(is_even, numbers))
print(evens)   # [2, 4, 6]
```

*Use map() when you want to apply a formula/action to everything.*

*Use filter() when you want to pick only some items that meet a condition.*

*Often, we use lambda with them for short, one-line functions.*

```python
#Task #01 Use map to convert ["ali", "sara", "waqas"] to uppercase.
names=["ali", "sara", "waqas"]
def up(x):
  x=x.upper()
  return x
uppercaselist=list(map(up,names))
print(uppercaselist)
```

```python
#Task #02 Use filter to keep numbers less than 10 from [5, 15, 8, 20].
nums=[5, 15, 8, 20]
def lessthan(x):
  if x<10:
    return x
lessthanlist=list(filter(lessthan,nums))
print(lessthanlist)

```
